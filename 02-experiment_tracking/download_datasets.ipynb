```python
import pandas as pd
import os
import urllib3 as urllib
```

```python
# Downloading data from the website if it's not present in the local directory already.

if not os.path.exists("green_tripdata_2021-01.csv"):
    dataset_url = (
        "https://s3.amazonaws.com/nyc-tlc/trip+data/green_tripdata_2021-01.csv"
    )
    urllib.request.urlretrieve(dataset_url, "green_tripdata_2021-01.csv")
```
